### D3m
Students which have taken 0 credits

```python
%%sql
SELECT *
FROM Students
WHERE Students.pNbr NOT IN (
        SELECT Students.pNbr
        FROM Students NATURAL JOIN TakenCourses
        GROUP BY Students.firstName, Students.lastName
        )
```

### D3n
Highest average grade student

```python
%%sql
SELECT Students.firstName, Students.lastName, AVG(grade) AS avg_grade
FROM Students NATURAL JOIN TakenCourses
GROUP BY Students.firstName, Students.lastName
ORDER BY AVG(grade) DESC
LIMIT 5
```

### D3o
List of all the students (person number) with there corresponding total credits taken

```python
%%sql
SELECT Students.pNbr, IFNULL(SUM(credits), 0) AS total_credits
FROM Students NATURAL LEFT JOIN TakenCourses NATURAL LEFT JOIN Courses
GROUP BY Students.pNbr
ORDER BY total_credits
```

### D3p
o, but replace person number with name

```python
%%sql
SELECT Students.firstName, Students.lastName, IFNULL(SUM(credits), 0) AS total_credits
FROM Students NATURAL LEFT JOIN TakenCourses NATURAL LEFT JOIN Courses
GROUP BY Students.pNbr
ORDER BY total_credits 
```

### D3q

```python
%%sql
SELECT *
FROM Students
WHERE (firstName, lastName) IN (
        SELECT firstName, lastName
        FROM Students
        GROUP BY firstName, lastName
        HAVING count(lastName) > 1 AND count(firstName) > 1)
```
